# Visualize data

```python
# Build a function to show categorical values disribution
def plot_bar(column):
    # temp df 
    temp = pd.DataFrame()
    # count categorical values
    temp['No_deposit'] = df[df['y'] == 'no'][column].value_counts()
    temp['Yes_deposit'] = df[df['y'] == 'yes'][column].value_counts()
    temp.plot(kind='bar')
    plt.xlabel(f'{column}')
    plt.ylabel('Number of clients')
    plt.title('Distribution of {} and deposit'.format(column))
    plt.show();
```

```python
#this function returns categorical variables
def return_categorical(df):
    categorical_columns = [column_name for column_name in df if df[column_name].dtype == 'object' and column_name != 'y']
    return categorical_columns

for col in return_categorical(df):
    plot_bar(col)
```

```python
# Build a function to show categorical values disribution
def plot_bar2(column):
    plt.subplot(1,3,1)
    # temp df 
    temp = pd.DataFrame()
    # count categorical values
    sns.distplot(df[column][df.y=='no'],  kde=False,label='No_deposit', color='red', bins=20)
    sns.distplot(df[column][df.y=='yes'],  kde=False, label='Yes_deposit', color='green', bins=20)
    #temp['No_deposit'] = df[df['y'] == 'no'][column].value_counts()
    #temp['Yes_deposit'] = df[df['y'] == 'yes'][column].value_counts()
    #temp.plot(kind='bar')
    plt.legend()
    plt.xlabel(f'{column}')
    plt.ylabel('Number of clients')
    plt.title('Distribution of {} and deposit'.format(column))
    
    plt.subplot(1,3,3)
    sns.boxplot(data=df, x="y",y=column)
    plt.title('Distribution of {} and deposit'.format(column))
    plt.show()

#fig1 = sns.distplot(df.ages, bins = 20)
#plt.show()
numeric_columns = ['ages','balances','durations','campaigns']

for col in numeric_columns:
    plot_bar2(col)
```

Durations data shows an interesting information that longer duration helps to secure term deposit subscription. Although this highly affecting the target result, it is however won't be included to the prediction model dataset due to the duration of calls is not known until the call has been made. 

# Data PreProcessing

```python
#Define X and y
X = df.drop(['days'], axis=1).drop(['durations'], axis=1).iloc[:,:-1] # column 'days' is also dropped as to be unnecessary, as the effect of 'months' is more important.
y = df.iloc[:,-1]
display(X)
display(y)
```

```python
#One Hot Encoding all the categorical data (nominal)

def onehotencode(col):
    return pd.concat([X,pd.get_dummies(X[col], prefix=col, drop_first=False)],axis=1).drop([col],axis=1)

# Data that need to be One Hot Encoded:
ohe_columns =['jobs','maritals','educations','contacts', 'months']

for col in ohe_columns:
    X = onehotencode(col)
```

```python
X.dtypes
```

```python
#Label Encoding the Ordinal Categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
le_col = ['defaults', 'housings', 'loans']
for col in le_col:
    X[col] = labelencoder_X.fit_transform (X[col])
display(X)
X.describe()
y = labelencoder_X.fit_transform(y)
y
```

```python
#Checking correlation between data
plt.figure(figsize=(30,15))
sns.heatmap(X.corr(),linewidths=.1, annot=True,cmap="YlGnBu")
```

```python
#splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split (X, y, test_size = 0.2,random_state = 4)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
## Correlation with independent Variable 
y2 = labelencoder_X.fit_transform(y)
y2 = pd.DataFrame(y2, columns=['y'])
x2 = pd.concat([X,y2], axis=1)
x3 = x2.drop(['y'], axis=1)
x3.corrwith(x2['y']).sort_values(ascending = False).plot.bar(figsize= (45, 15), title = 'Correlation with y', fontsize= 22, rot = 45, grid = True)
```

# Data Modelling

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

```python
results = pd.DataFrame(columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression
LRclassifier = LogisticRegression()
LRclassifier.fit(x_train, y_train)

# Predicting Test Set
y_pred = LRclassifier.predict(x_test)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=LRclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['Logistic Regression (Lasso)', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])
results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
## K-Nearest Neighbors (K-NN)
from sklearn.neighbors import KNeighborsClassifier
KNNclassifier = KNeighborsClassifier(n_neighbors=15, metric='minkowski', p= 2)
KNNclassifier.fit(x_train, y_train)

# Predicting Test Set
y_pred = KNNclassifier.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=KNNclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['K-Nearest Neighbors (minkowski)', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])

results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
## Naive Bayes
from sklearn.naive_bayes import GaussianNB
NBclassifier = GaussianNB()
NBclassifier.fit(x_train, y_train)

# Predicting Test Set
y_pred = NBclassifier.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=NBclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['Naive Bayes (Gaussian)', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])

results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
## Decision Tree
from sklearn.tree import DecisionTreeClassifier
DTclassifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
DTclassifier.fit(x_train, y_train)

#Predicting the best set result
y_pred = DTclassifier.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=DTclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['Decision Tree', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])
results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
## Random Forest Gini
from sklearn.ensemble import RandomForestClassifier
RFGclassifier = RandomForestClassifier(random_state = 0, n_estimators = 100,
                                    criterion = 'gini')
RFGclassifier.fit(x_train, y_train)

# Predicting Test Set
y_pred = RFGclassifier.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=RFGclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['Random Forest Gini', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])
results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
## Random Forest Entropy
from sklearn.ensemble import RandomForestClassifier
RFEclassifier = RandomForestClassifier(random_state = 0, n_estimators = 100,
                                    criterion = 'entropy')
RFEclassifier.fit(x_train, y_train)

# Predicting Test Set
y_pred = RFEclassifier.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="binary", pos_label=1)
rec = recall_score(y_test, y_pred, average="binary", pos_label=1)
f1 = f1_score(y_test, y_pred, average="binary", pos_label=1)
auc = roc_auc_score(y_test, y_pred)

accuracies = cross_val_score(estimator=RFEclassifier, X=x_train, y=y_train,cv=10)

model_results = pd.DataFrame([['Random Forest Entropy', acc, accuracies.mean(), prec, rec, f1, auc]],
               columns = ['Model', 'Accuracy', 'Accuracy(Cross_val)', 'Precision', 'Recall', 'F1 Score', 'AUC score'])

results = results.append(model_results, ignore_index = True)

cm= confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm, annot=True)

#Making the classification report
print(classification_report(y_test,y_pred))
```

```python
from sklearn import metrics

classifiers = [LRclassifier, KNNclassifier, NBclassifier, DTclassifier, RFGclassifier, RFEclassifier]
modelName = ['LR', 'KNN', 'NB', 'DT', 'RF-Gini', 'RF-Ent']

for cla in classifiers:
    y_pred = cla.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)
    auc = metrics.roc_auc_score(y_test, y_pred)
    plt.plot(fpr,tpr,label="AUC {}={:.3f}".format(modelName[classifiers.index(cla)],auc))


plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
